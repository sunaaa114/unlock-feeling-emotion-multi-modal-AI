```python
import os
import numpy as np
import re
# 디렉토리 경로 설정
video_directory = "D:\\emotion_dataset\\movie2"
video_files = os.listdir(video_directory)

# 파일 리스트를 숫자 기준으로 정렬
video_sorted_files = sorted(video_files)

for filename in video_sorted_files:
    # 파일 이름 출력 또는 원하는 작업 수행
    print(filename)

# 파일 이름에서 숫자를 추출하는 함수
def extract_numbers_from_filename(filename):
    numbers = re.findall(r'-(\d+)\.', filename)
    return [int(num) for num in numbers]

# 파일 리스트에서 숫자들만 추출하여 리스트로 만드는 함수
def extract_all_numbers(file_list):
    all_numbers = []
    for filename in file_list:
        numbers = extract_numbers_from_filename(filename)
        all_numbers.extend(numbers)
    return list(all_numbers)

# 파일 리스트에서 숫자들만 추출하여 리스트로 만듦
video_number_list = extract_all_numbers(video_files)

# 숫자 리스트 출력 또는 원하는 작업 수행
print(video_number_list)
```

# 텍스트 특징 추출

```python
tokenizer = BertTokenizer.from_pretrained("monologg/kobert")
model = BertModel.from_pretrained("monologg/kobert")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

df = pd.read_csv('textandemotion.csv')

def add_kobert_embeddings(df, text_column):
    embeddings_numpy = []
    for text in df[text_column]:
        tokens = tokenizer.tokenize(text)
        tokens = ["[CLS]"] + tokens + ["[SEP]"]
        token_ids = tokenizer.convert_tokens_to_ids(tokens)
        tokens_tensor = torch.tensor([token_ids])
        with torch.no_grad():
            model.to(device)
            tokens_tensor.to(device)
            outputs = model(tokens_tensor)
            sentence_embedding = outputs[0][:, 0, :].squeeze().cpu().numpy()
        embeddings_numpy.append(sentence_embedding)
    df['kobert_embeddings_numpy'] = embeddings_numpy
    return df

df = add_kobert_embeddings(df, 'Extracted Text')
```

## 동영상 파일을 오디오 파일로 변환하여 저장

```python
audio_directory = "D:\\emotion_dataset\\audio2"
```

```python
#음성 파일로 만들기
import moviepy.editor as mp
def video_to_wav(video_path, audio_path):
    # 동영상 파일을 읽어오기
    video = mp.VideoFileClip(video_path)

    # 음성 추출
    audio = video.audio

    # 추출된 음성을 WAV 파일로 저장
    audio.write_audiofile(audio_path, codec='pcm_s16le')

    # 메모리 해제
    audio.close()
    video.close()
    
    
for filename in video_sorted_files:
    if filename.endswith(".mp4") or filename.endswith(".m2ts"):
        video_path = os.path.join(video_directory, filename)
        r_file = filename.rsplit('.', 1)[0]
         # .wav 확장자를 가진 음성 파일 경로 (동영상 파일명과 확장자를 변경)
        audio_path = os.path.join(audio_directory, r_file + '.wav')

        # 동영상 파일을 WAV 파일로 변환
        video_to_wav(video_path, audio_path)
```

## 음성 특징 추출

```python
audio_df = pd.DataFrame(columns=['feature'])
i = 0
for filename in audio_sorted_files:
    if filename.endswith(".wav"):
        audio_path = os.path.join(audio_directory, filename)
        X, sample_rate = librosa.load(audio_path,duration=1,sr=22050*2)
        sample_rate = np.array(sample_rate)
        mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=25), axis=0)
        feature = mfccs
        audio_df.loc[i] = [feature]
        i=i+1
        
scaler = StandardScaler()

audio_df2 = pd.DataFrame(columns=['normalized'])
for k in range(len(audio_df)):
    features = audio_df.loc[k, 'feature'].reshape(1, -1)
    features_transposed = features.T

    # 정규화 적용
    scaled_features_transposed = scaler.fit_transform(features_transposed)

    # 다시 전치하여 shape을 (n_samples, n_features)로 변경
    scaled_features = scaled_features_transposed.T
    scaled_features = scaled_features.flatten()
    scaled_features.tolist()
    audio_df2.loc[k, 'normalized'] = scaled_features.tolist()

audio_3 = pd.DataFrame(audio_df2['normalized'].values.tolist())
audio_3 = audio_3.fillna(0)
```

```python
# 모든 행을 배열로 변환
rows_as_arrays = []
for index, row in audio_3.iterrows():
    row_array = np.array(row)
    rows_as_arrays.append(row_array)
```
